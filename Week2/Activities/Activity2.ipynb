# **Actividad: Creación de una Base de Datos en Python**

## ***Descripción***

A continuación debes crear una base de datos simple utilizando listas en Python. Una base de datos es una colección de datos organizados que se utilizan para almacenar y recuperar información. En este caso, simularemos una base de datos de estudiantes con información básica.

## ***Instrucciones***

1. Crea un programa en Python que permita a los usuarios realizar las siguientes acciones:

   a. Agregar un nuevo estudiante a la base de datos. Cada estudiante debe tener al menos un nombre, un número de identificación y una edad. Puedes utilizar una lista para almacenar estos datos.

   b. Mostrar la lista de estudiantes en la base de datos.

   c. Buscar un estudiante por su número de identificación y mostrar su información.

   d. Actualizar la información de un estudiante existente en la base de datos.

   e. Eliminar un estudiante de la base de datos.

2. Utiliza listas de Python para almacenar la información de los estudiantes. Por ejemplo, puedes crear una lista de diccionarios, donde cada diccionario representa un estudiante con sus datos.

3. Ofrece un menú interactivo que permita a los usuarios elegir las acciones que desean realizar en la base de datos.

4. Implementa la lógica necesaria para cada acción, como agregar, buscar, actualizar o eliminar estudiantes.

## ***Ejemplo de Base de Datos***

A continuación, se muestra un ejemplo de cómo podría verse una base de datos de estudiantes utilizando listas y diccionarios:

```python
base_de_datos = [
    {"nombre": "Juan", "id": 1, "edad": 20},
    {"nombre": "María", "id": 2, "edad": 22},
    {"nombre": "Carlos", "id": 3, "edad": 21}
]

Desarrolle el código en la siguiente lista.

------------------------------------------------------------------------------------------------------------

````python
# Base de datos de estudiantes
estudiantes_bd = [
    {"nombre": "Katie", "id": 7, "edad": 40},
    {"nombre": "Taylor", "id": 13, "edad": 34},
    {"nombre": "Jenna", "id": 8, "edad": 22}
]

#Funcion para mostrar el horario 
def mostrar_menu():
    #Muestra el menú de opciones al usuario.
    print("1. Agregar un nuevo estudiante")
    print("2. Mostrar la lista de estudiantes")
    print("3. Buscar un estudiante")
    print("4. Actualizar la información de un estudiante")
    print("5. Eliminar un estudiante")
    print("6. Salir")

#Funcion para agregar un nuevo estudiante
def agregar_estudiante():
    #Agrega un nuevo estudiante a la base de datos.
    nombre = input("Ingrese el nombre del estudiante: ")
    id = int(input("Ingrese el número de identificación del estudiante: "))
    edad = int(input("Ingrese la edad del estudiante: "))
    nuevo_estudiante = {"nombre": nombre, "id": id, "edad": edad}
    estudiantes_bd.append(nuevo_estudiante)
    print("El estudiante ha sido agregado exitosamente.\n")

#Funcion para mostrar los estudiantes
def mostrar_estudiantes():
    #Si no hay estudiantes en la bd 
    if not estudiantes_bd:
        #Mostrar este mensaje
        print("Ese estudiante no se encuentra en la base de datos")
    #Si hay estudiantes en la bd 
    else:
        #Mostrar estudiantes 
        print("\nLista de estudiantes:")
        for i, estudiante in enumerate(estudiantes_bd, start=1):
            print(f"{i}. {estudiante['nombre']} ({estudiante['id']}) - {estudiante['edad']} años")
        print(" ")

#Funcion para bscar un estudiante 
def buscar_estudiante():
    #Busca un estudiante por su número de identificación.
    id_buscar = int(input("Ingrese el número de identificación del estudiante a buscar: "))
    encontrado = False
    #por cada estudainte en la bd 
    for estudiante in estudiantes_bd:
         #Verificar si el id coincide con id ingresado por el usuario
        if estudiante["id"] == id_buscar:
            #Imprimir informacion del estudiante
            print("\nInformación del estudiante:")
            print(f"Nombre: {estudiante['nombre']}")
            print(f"Número de identificación: {estudiante['id']}")
            print(f"Edad: {estudiante['edad']} años")
            print(" ")
            encontrado = True 
            break
    #Si no fue encontrado 
    if not encontrado:
        #mostrar el siguiente mensaje
        print("No se encontró ningún estudiante con ese número de identificación \n")

#Funcion para actualizar informacion del estudinte 
def actualizar_estudiante():
    # Buscar al estudiante por su numero de identificacion
    id_actualizar = int(input("Ingrese el número de identificación del estudiante a actualizar: "))
    #por cada estudiante en la bd 
    for estudiante in estudiantes_bd:
        #Verificar si el id coincide con id ingresado por el usuario
        if estudiante["id"] == id_actualizar:
            nombre = input("Ingrese el nuevo nombre del estudiante: ")
            edad = int(input("Ingrese la nueva edad del estudiante: "))
            nuevo_id = int(input("Ingrese el nuevo número de identificación del estudiante: "))
           #Verificar si el nuevo id ya existe en la bd
            if any(estudiante["id"] == nuevo_id for estudiante in estudiantes_bd):
                #Si ya existe, mostar el siguiente mensaje
                print("El nuevo ID ya está siendo utilizado por otro estudiante. No se pudo actualizar el ID.\n")
            #Si no existe 
            else:
                #Actualizar informacion 
                estudiante["nombre"] = nombre
                estudiante["edad"] = edad
                estudiante["id"] = nuevo_id
                print("La información del estudiante ha sido actualizada exitosamente.\n")
            return
    print("No se encontró ningún estudiante con ese número de identificación.\n")

#Funcion para eliminar estudiante 
def eliminar_estudiante():
    #Buscar al estudiante por su numero de identificacion
    id_eliminar = int(input("Ingrese el número de identificación del estudiante a eliminar: "))
    #Por cada estudiante en la bd 
    for estudiante in estudiantes_bd:
        #Verificar si el id coincide con id ingresado por el usuario
        if estudiante["id"] == id_eliminar:
            #Eliminar el estudiante seleccionado 
            estudiantes_bd.remove(estudiante)
            print("El estudiante ha sido eliminado exitosamente.\n")
            return
    print("No se encontró ningún estudiante con ese número de identificación.\n")

while True:
    mostrar_menu()
    opcion = int(input("Elija una opción (1-6): "))
    if opcion == 1:
        agregar_estudiante()
    elif opcion == 2:
        mostrar_estudiantes()
    elif opcion == 3:
        buscar_estudiante()
    elif opcion == 4:
        actualizar_estudiante()
    elif opcion == 5:
        eliminar_estudiante()
    elif opcion == 6:
        print("¡Hasta luego!\n")
        break
    else:
        print("Opción invalida. Inténtelo de nuevo. \n")
````

# **Actividad: Rebanado de listas.**

Revisa la unidad 10,sección 5.3 aparece un resumen de los comandos para tomar secciones de una lista, aplica los comandos en la siguiente lista:


list_notas = [15, 21, 24, 40, 13, 14, 18, 17, 14.5]



````python
list_notas = [15, 21, 24, 40, 13, 14, 18, 17, 14.5]

#Start|Stop|Step
#Comienza en el inicio, llega al final con paso 2
print(list_notas[::2])

#Mustra desde el indice 1 hasta el indice 6 saltando de 2 en 2
print(list_notas[1:6:2])

#Sart|Stop
#Muestra el indice 1 hasta el indice 4
print(list_notas[1:5])

#Inverso 
#Devuelve la lista con orden inverso
print(list_notas[::-1])

#Star|
#Muestra desde el indice 2 hasta el final 
print(list_notas[2:])

#End|
#Muestra desde el inico hasta el indice 6
print(list_notas[:7])

#Muestra todo la lista
print(list_notas[:])

#Muestra los dos ultimos elementos 
print(list_notas[-2:])

#Muestra el indice 1 hasta el 0 (inicio) en orden inverso
print(list_notas[1::-1])
````

# **Actividad: Diferentes métodos usados en listas.**

De la unidad 10, sección 6.3 revise los métodos disponibles para usar sobre las listas.

Aplique los métodos disponibles sobre la lista mostrada en la siguiente celda. 

Nota: usando el método **sort()** ordene de mayor a menor y viceversa (Si es necesario revise documentación adicional relacionada sobre este método)


````python
list_notas = [15, 21, 24, 40, 13, 14, 18, 13, 17, 14.5]

#Index(x):
#Muestra el indice de la primera concurrencia del valor especificado
print(f"Index Method: {list_notas.index(14)} \n")

#Append(x):
#Añade un valor al final de la lista
list_notas.append(55)
print(f"Append Method: \n {list_notas} \n")

#Count(x):
#Muestra el numero de veces que valor apacere en la lista 
print(f"Count Method: {list_notas.count(13)} \n")

#Extend([x1, x2])
#Añade elementos de una lista o cualquier itrable (como tupla, cadena, numero) al final
lista_nueva = [1,2,8,13,7,3]
#Vamos agregar mas numeros al final de las lista 
list_notas.extend([22, 7, 5])
print(f"Extend Method con Numeros: \n {list_notas}")
#Vamos agregar los datos de una lista al final de list_notas 
list_notas.extend(lista_nueva)
print(f"Extend Method con Lista: \n {list_notas} \n")

#Insert(index, x):
#Agrega el valor especificado en la posicion especificada 
list_notas.insert(2, 80)
print(f"Insert Method: \n {list_notas} \n")

#Remove(X):
#Elimina la primera concurrencia del valor especificado
list_notas.remove(15)
print(f"Remove Method: \n {list_notas} \n")

#Pop(Index):
#Elimina el valor que se encuentra en el indice especificado
print(f"Pop Method: {list_notas.pop(1)}")
print(f"{list_notas} \n")

#Sort():
#Ordena los elementos de una lista de menor a mayor/ mayor a menor
list_notas.sort()
print(f"Sort Method: \n {list_notas}")
#Para ordenar los elementos de mayora menor usando sor()
list_notas.sort(reverse = True)
print(f"Sort Method (Reverse): \n {list_notas} \n")

#Reverse():
#Realiza un orden inverso de los elementos de la lista
list_notas.reverse()
print(f"Reverse Method: \n {list_notas}")
````

Genera un diccionario, sobre la información que deseas guardar como base de datos, desarrollo el código necesario para generar el archivo JSON, y también el código para cargar la información del archivo JSON.


````python
#Creamos el diccionario llamado Celebridades
celebridades = {
    1:{
        "nombre": "Taylor Swift",
        "edad": 34,
        "pais_nacimiento": "Estados Unidos"
    },
    2:{
        "nombre": "Lana del Rey",
        "edad": 38,
        "pais_nacimiento": "Estados Unidos"
    },
    3:{
        "nombre": "Billie Eilish",
        "edad": 22,
        "pais_nacimiento": "Estados Unidos"
    },
    4:{
        "nombre": "Jenna Ortega",
        "edad": 21,
        "pais_nacimiento": "Estados Unidos"
    },
    5:{
        "nombre": "Olivia Rodrigo",
        "edad": 20,
        "pais_nacimiento": "Estados Unidos"
    },
    6:{
        "nombre": "Katie McGrath",
        "edad": 40,
        "pais_nacimiento": "Irlanda"
    }
}
#Guardar el diccionario en un archivo JSON
#Se importa el modulo JSON
import json

#Abrirmos el archivo con modo de escritura ('w')
with open("celebridades.json","w") as f:
    #Se utiliza el metodo json.dump para guardar el diccionario en el archivo JSON
    json.dump(celebridades, f)
#Cerrar el archivo
f.close()

#Cargar la informacion del archivo JSON
#Abrimos el archivo con el modo lectura ('r')
with open("celebridades.json", "r") as f:
    #Se utiliza el metodo json.load para cargar el diccionario desde el archivo JSON
    celebridades_cargadas = json.load(f)
#Cerrar el archivo 
f.close()

#Imprimir el diccionario 
print(celebridades_cargadas)
````

#### **Diferencia entre lista y Tupla**

La principal diferencia es que las tuplas no se pueden modificar después de su creación y se representan con `()`. 
Miestras que las listas si se pueden modificar después de su creación y se representan con `[]`.

````python
#Creacion de Tupla 
tup = (1, 2, 3, 4, 5)

#Uso del Operador in

#Esta el numero 6 en la tupla 
print(6 in tup) 

#Esta el numero 3 en la tupla 
print(3 in tup)

#Usando IN con ciclo if 
#Si el numero tres se encuentra en la tupla 
if 3 in tup:
    print("La tupla contine el numero 3")
#Si no se encuentra en la tupla
else: 
    print("La tupla no contiene elnumero 3")
````

4. Sumar todo los elementos de esta lista bidimensional.

#### **Lista Bidemensional** 

Es una lista de lista, es decir en lugar de almacenar elementos individuales, almacena listas. 

````python
#Crear lista bidemensional 
list_2d = [[1, 3, 5], [2, 4, 6], [7, 9, 11], [8, 10, 12]]

#Inicializamos la variable suma en 0
suma = 0
#Iteramos sobre cada sublista de la lista bidemensional
for lista in list_2d:
    #Iteramos sobre cada elemento de la sublista 
    for e in lista:
        #Sumar el elemento a la variable suma 
        suma += e 

#Imprimir la suma de los elementos 
print(f"La suma de los elementos de la lista bidemensional es: {suma}")
````

1. Crear un diccionario.

3. Agregar un breve comentario sobre su funcionamiento.

````python
#Crear Diccionario 
# Crear un diccionario
cantantes = {
    1:{
        "nombre": "Taylor Swift",
        "edad": 34,
        "pais_nacimiento": "Estados Unidos"
    },
     2:{
        "nombre": "Lana del Rey",
        "edad": 38,
        "pais_nacimiento": "Estados Unidos"
    },
    3:{
        "nombre": "Billie Eilish",
        "edad": 22,
        "pais_nacimiento": "Estados Unidos"
    }
}

# Método get(): 
# Obtiene el valor asociado a una clave o devuelve un valor predeterminado si la clave no existe
nombre_cantante = cantantes.get(1).get("nombre")
print(f"El nombre de la cantante con ID 1 es: \n {nombre_cantante} \n")

#Método keys(): 
#Devuelve una vista de todas las claves del diccionario
claves = list(cantantes.keys())
print(f"Las claves del diccionario son: \n {claves} \n")

#Método values(): 
#Devuelve una vista de los valores del diccionario
#singers = list(cantantes.values())
singers = list(cantantes.values())
print(f"Las cantantes son: \n {singers} \n")

#Método items(): 
#Devuelve una vista de las clave-valor del diccionario
clave_cantante = list(celebridades.items())
print(f"Las cantantes son: \n {clave_cantante} \n")

#Método pop():
#Elimina y devuelve el valor asociado a la clave especificada
eliminar = cantantes.pop(3)
print(f"El nombre de la cantante con ID 1 es: \n {eliminar} \n")

#Método clear():
#Elimina todos los elementos del diccionario.
cantantes.clear()
print("Diccionario vacio", cantantes)
````
