## **Numpy Array vs Python List**
-------------------------
---------------------------


| Feature        | Python List | NumPy Array |
| --------------| ---------- | ----------- |
| **Storage**       | Dynamic    | Fixed       |
| **Functionality** | General-purpose | Mathematical and scientific operations |
| **Speed**         | Slower due to type-checking and overhead | Faster due to vectorized operations and no type-checking |
| **Memory**       | More due to overhead | Optimized and efficient storage |

### ***Storage Comparison*** 
------------------

```python
# import the numpy library and give it the alias "np"
import numpy as np 
# import the sys module, which contains functions related to the Python runtime environment
import sys 

# Create a list of integers from 0 to 999 (exclusive) and assign it to the variable "b"
b = range(1000)

# Print the amount of memory used to store the list "b" in bytes.
# The syntax `sys.getsizeof(5)` gets the size of an integer object in bytes,
# and we multiply it by the length of the list to get the total memory usage.
print(f'List Storage: {sys.getsizeof(5)*len(b)}')

# Create a numpy array of integers from 0 to 999 (exclusive) and assign it to the variable "c"
c = np.arange(1000)

# Print the amount of memory used to store the numpy array "c" in bytes.
# The `size` attribute of a numpy array returns the number of elements in the array, 
# and the `itemsize` attribute returns the size of each element in bytes.
print(f"Numpy Array Storage: {c.size*c.itemsize}")
```

### ***Speed Comparison***
----------------

```python
# import the numpy library and give it the alias "np"
import numpy as np 
# import the time module, which contains functions for measuring elapsed time
import time 
# import the sys module, which contains functions related to the Python runtime environment
import sys 

# Set the size of the lists and arrays to 100,000
size = 100000

# Create two lists of integers from 0 to 99,999 (exclusive) and assign them to the variables "L1" and "L2"
L1 = range(size)
L2 = range(size)

# Create two numpy arrays of integers from 0 to 99,999 (exclusive) and assign them to the variables "A1" and "A2"
A1 = np.arange(size)
A2 = np.arange(size)

# Record the current time
start = time.time()
# Use a list comprehension to create a new list that contains the sum of each pair of corresponding elements in "L1" and "L2"
# Assign the resulting list to the variable "result"
result = [(x+y) for x, y in zip(L1, L2)]
# Calculate and print the time in milliseconds that it took to execute the list comprehension
print(f"Python List took:{(time.time() - start)*1000}ms")

# Record the current time
start = time.time()
# Add the numpy arrays "A1" and "A2" element-wise and assign the result to the variable "result"
result = A1+A2
# Calculate and print the time in milliseconds that it took to execute the numpy array addition
print(f"Numpy Array took:{(time.time() - start)*1000}ms")
```

### ***Creating Arrays, Finding size and shape of Arrays and Arrange functions***
-----------------------------

```python
# import the numpy library and give it the alias "np"
import numpy as np 

# Create a 2-dimensional numpy array with integers and assign it to the variable "a"
a = np.array([[1, 2], 
              [3, 4], 
              [5, 6]])

# Print the array "a"
print("Array: ")
print(a)

# Print the number of array dimensions (2, in this case)
print("\n------------------------------------")
print("The number of array dimensions: ")
print(a.ndim)

# Print the size of each array element in bytes
print("\n------------------------------------")
print("The size of each array element in bytes: ")
print(a.itemsize)

# Print the shape of the array (3 rows and 2 columns, in this case)
print("\n------------------------------------")
print("The shape of the array: ")
print(a.shape)

# Create a 2-dimensional numpy array with floating-point numbers and assign it to the variable "a2"
a2 = np.array([[1, 2], 
              [3, 4], 
              [5, 6]], dtype=np.float64)

# Print the array "a2"
print("\n------------------------------------")
print("Array of Floats: ")
print(a2)

# Print the number of array dimensions (2, in this case)
print("\n------------------------------------")
print("The number of array dimensions: ")
print(a2.ndim)

# Print the size of each array element in bytes
print("\n------------------------------------")
print("The size of each array element in bytes: ")
print(a2.itemsize)

# Print the shape of the array (3 rows and 2 columns, in this case)
print("\n------------------------------------")
print("The shape of the array: ")
print(a2.shape)

# Create a 2-dimensional numpy array with complex numbers and assign it to the variable "a3"
a3 = np.array([[1, 2], 
              [3, 4], 
              [5, 6]], dtype=np.complex128)

# Print the array "a3"
print("\n------------------------------------")
print("Array of Complex Numbers: ")
print(a3)

# Create a 2-dimensional numpy array of zeros and assign it to the variable "a0"
a0 = np.zeros((3,4))

# Print the array "a0"
print("\n------------------------------------")
print("Array of Zeros: ")
print(a0)

# Create a 2-dimensional numpy array of ones and assign it to the variable "a1"
a1 = np.ones((3,4))

# Print the array "a1"
print("\n------------------------------------")
print("Array of Ones: ")
print(a1)

# Create a numpy array with integers from 0 to 4 (exclusive) and assign it to the variable "nl"
nl = np.arange(5)

# Print the array "nl"
print("\n------------------------------------")
print("Array: ")
print(nl)
```

### ***Strings Functions***
--------------------------

```python
# import the numpy library and give it the alias "np"
import numpy as np  

# Concatenate the strings 'Hello' and 'Hi' with the strings 'abc' and 'xyz' using the `add` function
# from the `np.char` module and print the result
print("Concatenation Example: ")
print(np.char.add(['Hello', 'Hi'], ['abc', 'xyz']))

# Multiply the string 'Hello ' by 3 using the `multiply` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Multiply Example: ")
print(np.char.multiply('Hello ', 3))

# Center the string 'Hello' in a field of width 20 with the fill character '-' using the `center` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Center Example: ")
print(np.char.center('Hello', 20, fillchar='-'))

# Capitalize the first letter of the string 'hello universe' using the `capitalize` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Capitalize Example: ")
print(np.char.capitalize('hello universe'))

# Capitalize the first letter of each word in the string 'hello universe' using the `title` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Tittle Example: ")
print(np.char.title('hello universe'))

# Convert all the characters in the strings 'HELLO' and 'UNIVERSE' to lowercase using the `lower` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Lower Examples: ")
print(np.char.lower(['HELLO', 'UNIVERSE']))
print(np.char.lower('HELLO'))

# Convert all the characters in the strings 'python' and 'universe' to uppercase using the `upper` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Upper Examples: ")
print(np.char.upper(['python', 'universe']))
print(np.char.upper('python is easy'))

# Split the string 'python is easy, you should learned it' into a list of words using the `split` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Split Example: ")
print(np.char.split('python is easy, you should learned it'))

# Split the string 'python is easy\nyou should learned it' into a list of lines using the `splitlines` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Splitlines Example: ")
print(np.char.splitlines('python is easy\nyou should learned it'))

# Remove the first and last characters (assumed to be 'a') from the strings 'anit', 'admind', and 'nina' using 
# the `strip` function from the `np.char` module and print the result
print("\n------------------------------------")
print("strip Example: ")
print(np.char.strip(['anit', 'admind', 'nina'], 'a'))

# Join the strings 'dmy' and 'ymd' with the string ':' using the `join` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Join Example: ")
print(np.char.join([":", "-"], ['dmy', 'ymd']))

# Replace the string 'is' with the string 'was' in the string 'He is a good dancer' using the `replace` function
# from the `np.char` module and print the result
print("\n------------------------------------")
print("Replace Example: ")
print(np.char.replace('He is a good dancer', 'is', 'was'))
```

### ***Array Manipulation***
--------------------------

```python
# import the numpy library and give it the alias "np"
import numpy as np 

# Create a 1-dimensional numpy array with integers from 0 to 8 (exclusive) and assign it to the variable "a"
a =np.arange(9)

# Print the array "a"
print("Original Array: ")
print(a)

# Reshape the array "a" into a 2-dimensional array with 3 rows and 3 columns and assign it to the variable "b"
b = a.reshape(3, 3)

# Print the array "b"
print("\n------------------------------------")
print('Modify Array: ')
print(b)

# Flatten the array "b" into a 1-dimensional array and print it
print("\n------------------------------------")
print('Flatten Array: ')
print(b.flatten())

# Flatten the array "b" into a 1-dimensional array in Fortran order (column-major) and print it
print("\n------------------------------------")
print('Flatten Array Fortran: ')
print(b.flatten(order = "F"))

# Create a 2-dimensional numpy array with integers from 0 to 11 (exclusive) and assign it to the variable "a"
a = np.arange(12).reshape(4, 3)

# Print the array "a"
print("\n------------------------------------")
print("New Array by Arange & reshaping: ")
print(a)

# Transpose the array "a" and print it
print("\n------------------------------------")
print("Transpose Array: ")
print(np.transpose(a))

# Create a 2-dimensional numpy array with integers from 0 to 7 (exclusive) and assign it to the variable "b"
b = np.arange(8).reshape(2, 4)

# Print the array "b"
print("\n------------------------------------")
print("New Array by Arange & reshaping: ")
print(b)

# Reshape the array "b" into a 3-dimensional array with 2 slices along the first dimension, 
# 2 slices along the second dimension, and 2 slices along the third dimension, and print it
print("\n------------------------------------")
c = b.reshape(2, 2, 2)
print("New Array by reshaping: ")
print(c)

# Roll the third axis of the array "c" to the first position and print it
print("\n------------------------------------")
print("New Array Rollaxis: ")
print(np.rollaxis(c, 2, 1))

# Roll the third axis of the array "c" to the first position, which is the default behavior of the function, and print it
print("\n------------------------------------")
print("New Array Rollaxis: ")
print(np.rollaxis(c, 2))

# Roll the second axis of the array "c" to the first position and print it
print("\n------------------------------------")
print("New Array Rollaxis: ")
print(np.rollaxis(c, 1))
```

```python
# import the numpy library and give it the alias "np"
import numpy as np  

# Create a 2-dimensional numpy array with integers from 0 to 8 (exclusive) and assign it to the variable "a"
a = np.arange(9).reshape(3, 3)

# Print the array "a"
print("Array A: ")
print(a)

# Create a 1-dimensional numpy array with integers 10, 12, and 12 and assign it to the variable "b"
b = np.array([10, 12, 12])

# Print the array "b"
print("\n------------------------------------")
print("Array B: ")
print(b)

# Add the arrays "a" and "b" element-wise and print the result
print("\n------------------------------------")
print('Suma de Array A+B')
print(np.add(a, b))

# Subtract the arrays "a" and "b" element-wise and print the result
print("\n------------------------------------")
print('Resta de Array A+B')
print(np.subtract(a, b))

# Multiply the arrays "a" and "b" element-wise and print the result
print("\n------------------------------------")
print('Multiplicacion de Array A+B')
print(np.multiply(a, b))

# Divide the array "a" by the array "b" element-wise and print the result
print("\n------------------------------------")
print('Division de Array A+B')
print(np.divide(a, b))
```

### ***Iterating Over Arrays***
-----------------------------

```python
# import the numpy library and give it the alias "np"
import numpy as np  

# Create a 2-dimensional numpy array with integers from 0 to 40 (exclusive) in steps of 5, 
# and reshape it into a 3x3 array
a = np.arange(0, 45, 5)
a = a.reshape(3, 3)

# Print the array "a"
print("Array: ")
print(a)

# Iterate over the elements of the array "a" using the `np.nditer` function and print each element
print("\n------------------------------------")
print("Print Array with iteration: ")
for x in np.nditer(a):
    print(x)

# Iterate over the elements of the array "a" using the `np.nditer` function 
# with the 'C' order (row-major order) and print each element
print("\n------------------------------------")
print("Print Array with iteration order with C-Style: ")
for x in np.nditer(a, order='C'):
    print(x)

# Iterate over the elements of the array "a" using the `np.nditer` function 
# with the 'F' order (column-major order) and print each element
print("\n------------------------------------")
print("Print Array with iteration order with F-Style: ")
for x in np.nditer(a, order='F'):
    print(x)
```

### ***Joining Array***

```python
# import the numpy library and give it the alias "np"
import numpy as np  

# Create a 2-dimensional numpy array "a" with integers 1, 2, 3, and 4
a = np.array([[1, 2], [3, 4]])

# Print the array "a"
print('First Array: ')
print(a)

# Create a 2-dimensional numpy array "b" with integers 5, 6, 7, and 8
b = np.array([[5, 6], [7, 8]])

# Print the array "b"
print("\n------------------------------------")
print('Second Array: ')
print(b)

# Join the arrays "a" and "b" along the vertical axis (axis=0) using the `np.concatenate` 
print("\n------------------------------------")
print('Joining the two Arrays along axis 0:')
print(np.concatenate((a, b)))

# Join the arrays "a" and "b" along the horizontal axis (axis=1) using the `np.concatenate` 
print("\n------------------------------------")
print('Joining the two Arrays along axis 1:')
print(np.concatenate((a, b), axis=1))
```

```python
"""
    Find the total number of locations of missing values in the array 
"""
# import the numpy library and give it the alias "np"
import numpy as np

# Create a 2-dimensional numpy array "z" with shape (10, 10) and random values between 0 and 1 
# using the `np.random.rand` function
z = np.random.rand(10, 10)

# Select 5 random row indices and 5 random column indices using the `np.random.randint` function, 
# and set the corresponding elements in the array "z" to `np.nan` using slicing and assignment
z[np.random.randint(10, size=5), np.random.randint(10, size=5)] = np.nan

# Print the array "z"
print(z)


print('\n-------------------------------------------------------------------------------')
# Compute the number of missing values in the array "z" using the `np.isnan` function and the `sum` function
print("Total numbers of missing values: \n", np.isnan(z).sum())

# Compute the indices of the missing values in the array "z" using the `np.isnan` function 
# and the `argwhere` function
print('Indexes of missing values: \n', np.argwhere(np.isnan(z)))
```
