# **Actividad: Importación de módulos y su uso.**


## ***Instrucciones***

1. Importe los módulos que te permitan trabajar con fechas y tiempo (modulo datetime).

2. Use los diferentes métodos relacinados a operaciones con fechas y hora.

3. En el siguiente apartado revisa cada una las fechas contenidas en la lista e imprima el número de semanas y los dias restantes para llegar a esa fecha.

```python
# Importar Modulo 
import datetime as dt

# Métodos relacionados a operaciones con fecha y hora
# Fecha y tiempo actual
now = dt.datetime.now()
print(f"Fecha y tiempo actual: \n{now}")

# Fecha actual 
today = dt.date.today()
print(f"\nFecha del dia de hoy: \n{today}")

# Hora actual
tiempo = dt.datetime.now().time()
print(f"\nTiempo Actual: \n{tiempo}")
print("--------------------------------\n")


# Fución para calcular el tiempo faltante 
def calcular_tiempo(fecha):
    # Definir el formato de fecha
    format = '%m/%d/%Y'
    # Obtener la fecha actual
    fecha_actual =  dt.datetime.now()
    # Convertir la fecha a datetime
    fecha_proxima = dt.datetime.strptime(fecha, format)
    
    # Calcular la diferencias tiempos 
    diferencia = fecha_proxima - fecha_actual
    semanas = diferencia.days//7
    dias = diferencia.days%7
    
    # Devolver el resultado
    return semanas, dias

# Lista 
lista_de_fechas = ['04/01/2024', '01/28/2024', '03/16/2024', '05/24/2024', '07/30/2025']

# Realizar el calculo de la funcion por cada fecha en la lista 
for fecha in lista_de_fechas:
    semanas, dias = calcular_tiempo(fecha)
    print(f'Fecha: {fecha}\n Semanas: {semanas} \n Días restantes: {dias}\n')
```

# **Actividad: Manejo de la biblioteca Pandas.**


## ***Instrucciones***


1. Crear una serie de pandas, donde su indice este compuesto por fechas y en columna principal almace los valoras cargados en una lista.

2. Graficar los valores de la lista en función de la fecha.


```python
# Librerías 
import pandas as pd 
import matplotlib.pyplot as plt 

# Lista 
precios = [140, 130, 120, 115, 160, 110, 170, 150, 157, 135]

# Fechas 
dates = pd.date_range("2024-03-01", "2024-03-10", freq='D')

# Crear una Serie
s = pd.Series(precios, index= dates)

# Graficar la Serie
s.plot()

# Configurar Labels y Titulo
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.title('Gráfico de Precios en función de las Fechas')


# Mostrar la gráfica
plt.show()
```

# **Actividad: Operaciones con Data Frames**


## **Instrucciones**


1. Dado el siguiente diccionario construya un objeto DataFrame.
2. Filtre los resultados por año de publicación.
3. Obtenga algunas columnas basado en que contenga alguna palabra en el nombre de la publicacion.


```python
# Librerías 
import pandas as pd 
from IPython.core.display import HTML

# Diccionario 
publicaciones_literarias = {
    1605: "Don Quijote de la Mancha - Miguel de Cervantes",
    1865: "Alicia en el país de las maravillas - Lewis Carroll",
    1949: "1984 - George Orwell",
    1954: "El Señor de los Anillos - J.R.R. Tolkien",
    1997: "Harry Potter y la piedra filosofal - J.K. Rowling",
    2003: "El Código Da Vinci - Dan Brown"
}

# Crear DataFrame
df = pd.DataFrame(publicaciones_literarias.items(), columns=['Año', 'Publicación'] )
# Mostrar DataFrame 
print("DataFrame:")
display(HTML(df.to_html()))

# Filtrar por año de publicación, Donde el año se mayor 1950
year_df = df[df['Año'] < 1950]
# Mostrar resultado  
print('\n-------------------------------------------------------')
print('Filtrado por Año: ')
display(HTML(year_df.to_html()))

# Filtrar conteniendo alguna palabra en el nombre de la publicacion
palabra = df[df['Publicación'].str.contains('Vinci')]
# Mostrar resultado 
print('\n-------------------------------------------------------')
print('Filtrado por palabra: ')
display(HTML(palabra.to_html()))
```
