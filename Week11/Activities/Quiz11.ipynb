````python
# DEFINIR RUTA DEL DIRECTORIO DONDE SE ENCUENTRAN LOS DATOS 
ruta_archivo = r'C:\Users\cabre\Documents\Cursos\SIC\Week11\Dataset\data_sales.csv'

# LEER EL ARCHIVO CVS 
df_sales = pd.read_csv(ruta_archivo, header='infer', na_values=[' '])

# EXPLORAR LOS DATOS 
print(f"Dimensiones de los datos: \n {df_sales.shape} \n") 
print("Primera 5 filas del Dataset:")
display(HTML(df_sales.head(5).to_html()))
````

````python
# AGRUPAR POR REGION Y CALCULAR MEDIA 
unitprice_media_region = df_sales.groupby('Region')['Unit Price'].mean()

# CONVERTIR RESULTADS EN DF
df_unitprice_media_region = pd.DataFrame({'Region': unitprice_media_region.index, 'Unit Price Mean': unitprice_media_region.values})

# MOSTRAR 
display(HTML(df_unitprice_media_region .to_html()))
````

````python
# METODO PIVOT_TABLE PARA CALCULAR MEDIA POR REGION 
unitprice_media_pv = df_sales.pivot_table(index='Region', values='Unit Price', aggfunc='mean')

# CONVERTIR RESULTADS EN DF
df_unitprice_media_pv = pd.DataFrame(unitprice_media_pv.reset_index())
df_unitprice_media_pv = df_unitprice_media_pv.rename(columns={'Unit Price': 'Unit Price Mean'})

# MOSTRAR 
display(HTML(df_unitprice_media_pv.to_html()))
````

1.   Haz una tabla de frecuencia de 'yearOfStart' y visual칤zala por a침o.
- Ordenar por a침o.
- Dibuja un gr치fico de l칤neas desde 1997 hasta 2014. <= Pista: plt.xlim()

### **Respuesta:**

````python
# IMPORTAR LIBRERIAS 
import os
import pandas as pd
from IPython.core.display import HTML
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
````

````python
# DEFINIR RUTA DEL DIRECTORIO DONDE SE ENCUENTRAN LOS DATOS 
ruta_archivo = r'C:\Users\cabre\Documents\Cursos\SIC\Week11\Dataset\data_coffeeshop.csv'

# LEER EL ARCHIVO CVS 
df_Coffee = pd.read_csv(ruta_archivo, header='infer', na_values=[' '])

# EXPLORAR LOS DATOS 
print(f"Dimensiones de los datos: \n {df_Coffee.shape} \n") 
print("Primera 5 filas del Dataset:")
display(HTML(df_Coffee.head(5).to_html()))
````

````python
# CREACION DE TABLA DE FRECUENCIA 
tabla_frecuencia = df_Coffee['Year of Start'].value_counts()

# CONVERTIR TABLA DE FRECUENCIA A UN DF
df_frecuencia = pd.DataFrame({'Year of Start': tabla_frecuencia.index, 'Frecuencia': tabla_frecuencia.values})
    
# ORDENAR POR YEAR 
df_frecuencia = df_frecuencia.sort_values(by='Year of Start')

# MOSTRAR TABLA 
display(HTML(df_frecuencia.to_html(index=False)))
````

````python
# FILTAR LA TABLA DE FRECUENCIA CON LOS YEARS INDICADOS PARA LA GRAFICA
df_frecuencia_filt = df_frecuencia[(df_frecuencia['Year of Start'] >= 1997) & (df_frecuencia['Year of Start'] <= 2014)]

# CREAR GAFICA 
plt.figure(figsize=(10, 6)) 
sns.set(style="darkgrid")  
sns.lineplot(data=df_frecuencia_filt, x='Year of Start', y='Frecuencia', marker='o', color="#102C57", linewidth=2.5)

# ETIQUETAS Y TITULOS 
plt.xlabel('A침o', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)
plt.title('Gr치fico de l칤nea de la tabla de frecuencia (1997-2014)', fontsize=16)

# MOSTRAR
plt.show()
````
