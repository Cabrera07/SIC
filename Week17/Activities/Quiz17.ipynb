## ***Q 07-01***

### **¿Cuál de las siguientes afirmaciones sobre la minería de textos es incorrecta?**
1. Puedes conocer la reacción ante un grupo específico.
2. Se puede obtener información sobre oportunidades de ventas.
3. Se pueden establecer estrategias competitivas monitoreando otras marcas.
4. El mismo método se puede aplicar a varios idiomas.

### **Respuesta:**

***La afirmación sobre la minería de textos que esta incorrecta es la opción 4:***

- El mismo método se puede aplicar a varios idiomas.

**Explicación:**

Aunque algunos métodos de minería de textos pueden ser aplicables a varios idiomas, generalmente se requieren adaptaciones específicas para cada idioma debido a las diferencias en la estructura gramatical, el vocabulario, la semántica y otros factores lingüísticos. Por ejemplo, técnicas de procesamiento de lenguaje natural (NLP) como el análisis de sentimientos o la extracción de entidades nombradas necesitan ser ajustadas y entrenadas específicamente para cada idioma para obtener resultados precisos.

## ***Q 07-02***

### **¿Cuál de las siguientes afirmaciones no es cierta sobre el uso de las redes sociales?**
1. Es posible saber de cuántos grupos está compuesta la red.
2. Es posible conocer clientes influyentes.
3. Es posible ver el cambio a lo largo del tiempo.
4. Es posible saber si los clientes se irán la próxima vez.

### **Respuesta:**

***La afirmación sobre el uso de las redes sociales que esta incorrecta es la opción 4:***

- Es posible saber si los clientes se irán la próxima vez.

## ***Q 07-03***

### **¿Qué significa el término separar una raíz de una palabra cuya forma ha sido modificada para extraerla? ¿Cuál es la palabra que es objeto de análisis de morfemas?**

### **Respuesta:**

El término "separar una raíz de una palabra cuya forma ha sido modificada para extraerla" se refiere a la lematización. Este proceso reduce una palabra a su forma base o raíz (lema), considerando el contexto y la parte del discurso.

- **Lematización:** Proceso de convertir palabras inflectadas a su forma base.
  - **Ejemplo:** "corriendo", "corrí", "corre" se convierten en "correr".

La palabra que es objeto de análisis de morfemas es un morfema, la unidad mínima de significado en una lengua.

- **Análisis de Morfemas:** Identificación y separación de morfemas en una palabra.
  - **Ejemplo:** En "incompleto", los morfemas son "in-" (negación) y "completo" (raíz).

## ***Q 07-04***

### **¿Qué significa el término un conjunto de materiales que pueden mostrar los aspectos esenciales del lenguaje como material de investigación requerido en cada campo de la investigación del lenguaje?**

### **Respuesta:**

**Corpus:** Un corpus es un conjunto estructurado de textos recopilados de manera sistemática que se utiliza para estudiar y analizar aspectos del lenguaje en diversos campos de la investigación lingüística. Estos textos pueden incluir ejemplos de uso real del lenguaje, lo que permite a los investigadores examinar patrones, tendencias y características esenciales del lenguaje.

## ***Q 07-05***

### **Explique sobre TF (Frecuencia de términos) y escriba cómo calcular TF (fórmula).**

### **Respuesta:**

**TF (Frecuencia de Términos)**

Frecuencia de Términos (TF) es una medida que se utiliza en la minería de textos y recuperación de información para evaluar la importancia de una palabra específica dentro de un documento en relación con el documento mismo.

Es decir mide cuántas veces aparece un término en un documento en comparación con el total de términos en ese documento.

**Cálculo de TF:**

$TF(t, d) = f(t, d) / Nd$

***Donde:***

- TF(t, d) es la frecuencia del término $t$ en el documento $d$.
- $f(t, d)$ es el número de veces que el término $t$ aparece en el documento $d$.
- $Nd$ es el número total de términos en el documento $d$.
