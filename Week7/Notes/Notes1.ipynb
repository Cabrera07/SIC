### ***Series***
----------------------------------
- A Series is a one-dimensional labeled array capable of holding any data type.

- It can be created from lists, NumPy arrays, or other Series.

- Each element in a Series has an associated label, called an index.

----------------------------------
### ***DataFrame***
----------------------------------
- A DataFrame is a two-dimensional structure, similar to a database table or an Excel spreadsheet.

- It can contain data of different types in each column.

- It can be created from dictionaries, lists of Series, or NumPy arrays.

```python
'''
    Ejemplo Modificando el index
'''

# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Create a list of integers
arr = [0, 1, 2, 3, 4]

# Create a pandas Series object from the list of integers
s1 = pd.Series(arr)

# Print the Series object to the console with the default integer index
print(f'Default Index: \n{s1}\n')
print('---------------------------')

# Set the index of the Series object to a list of strings
s1.index = ['a', 'b', 'c', 'd', 'e']

# Print the Series object to the console with the custom string index
print(f'Custom Index: \n{s1}')
```

```python
'''
    Ejemplo Con diferentes Visualizaciones, utilizando slicing 
'''
# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Create a list of integers
arr = [0, 1, 2, 3, 4]

# Create a pandas Series object from the list of integers, with a custom string index
s1 = pd.Series(arr, index=['a', 'b', 'c', 'd', 'e'])

# Slice the Series object to get the first three elements
a = s1[:3]

# Print the sliced Series object to the console
print(f'The first three elements: \n{a}\n')
print('---------------------------')

# Slice the Series object to get all elements except for the last two
b = s1[:-2]

# Print the sliced Series object to the console
print(f'All elements except for the last two: \n{b}\n')
print('---------------------------')

# Slice the Series object to get all elements except for the last one
c = s1[:-1]

# Print the sliced Series object to the console
print(f'All elements except for the last one: \n{a}\n')
print('---------------------------')

# Slice the Series object to get the last two elements
d = s1[-2:]

# Print the sliced Series object to the console
print(f'The last two elements: \n{d}')
```

```python
'''
    Ejemplo Con diferentes Visualizaciones, concatenando series y quitando elementos 
'''
# Create a list of integers
arr = [0, 1, 2, 3, 4]

# Create a pandas Series object from the list of integers
s1 = pd.Series(arr)

# Set the index of the Series object to a list of strings
s1.index = ['a', 'b', 'c', 'd', 'e']

# Create a dictionary where the keys are strings and the values are integers
d = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

# Create a pandas Series object from the dictionary
s4 = pd.Series(d)

# Concatenate the two Series objects along the horizontal axis
s5 = pd.concat([s1, s4])

# Print the concatenated Series object to the console
print(f'Concatenate Series: \n{s5}\n')
print('---------------------------')

# Drop the 'e' index label from the concatenated Series object
e = s5.drop('e')

# Print the dropped Series object to the console
print(f"Drop 'e': \n{e}\n")
print('---------------------------')

# Print the original concatenated Series object to the console to show 
# that it was not modified by the drop() method
print(f'Concatenate Series: \n{e}\n')
```

```python
'''
    Ejemplos b√°sicos utilizando operaciones en Panda Series 
'''
# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Create two lists of integers
arr1 = [0, 1, 2, 3, 4, 5, 7]
arr2 = [6, 7, 8, 9, 5]

# Create a pandas Series object from the second list of integers
s6 = pd.Series(arr2)

# Print the Series object to the console
print(f"Serie1: \n{s6}\n")
print('---------------------------')

# Create a pandas Series object from the first list of integers
s7 = pd.Series(arr1)

# Print the Series object to the console
print(f"Serie2: \n{s7}\n")
print('---------------------------')

# Add the two Series objects and print the result to the console
suma = s6.add(s7)
print(f"Suma de Serie1 y Serie2: \n{suma}\n")
print('---------------------------')

# Subtract the second Series object from the first Series object and print the result to the console
resta = s6.sub(s7)
print(f"Resta de Serie1 y Serie2: \n{resta}\n")
print('---------------------------')

# Multiply the two Series objects and print the result to the console
multiplicar = s6.mul(s7)
print(f"Multiplicacion de Serie1 y Serie2: \n{multiplicar}\n")
print('---------------------------')

# Divide the second Series object by the first Series object and print the result to the console
dividir = s6.div(s7)
print(f"Division de Serie1 y Serie2: \n{dividir}\n")
print('---------------------------')

# Calculate the median of the second Series object and print it to the console
mediana = s6.median()
print(f"Mediana de Serie1: \n{mediana}\n")
print('---------------------------')

# Calculate the maximum value of the second Series object and print it to the console
maximo = s6.max()
print(f"Valor maximo de Serie1: \n{maximo}\n")
print('---------------------------')

# Calculate the minimum value of the second Series object and print it to the console
minimo = s6.min()
print(f"Valor minimo de Serie1: \n{minimo}\n")

```

```python
'''
    Ejemplos b√°sicos utilizando operaciones de b√°sicas de String en Panda Series 
'''

# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Create a pandas Series object with strings
string = pd.Series(['A', 'C', 'D', 'Aaa', 'BaCa', np.nan, 'CBD', 'cow', 'owl'])

# Convert all the strings in the Series object to lowercase and print the result to the console
low = string.str.lower()
print(f'Example Lowercase: \n{low}')
print("-------------------")

# Convert all the strings in the Series object to uppercase and print the result to the console
up = string.str.upper()
print(f'\nExample Uppercase: \n{up}')
```

```python
'''
    Ejemplos realizando operaciones con Missing Values en Panda DataFrame 
    Reemplazanndo the missing values con valores indicados 
'''
# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Import the IPython display module to display the DataFrame object in a more readable format
from IPython.core.display import HTML

# Define a dictionary with data for the DataFrame
data = {
    'Animal': 
        ['Cat', 'Cat', 'Snake', 'Dog', 'Dog', 'Cat', 'Snake', 'Cat', 'Dog', 'Dog'],
    'Age': 
        [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'Visits':
        [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'Priority':
        ['Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'No']
}

# Define a list of strings for the row labels
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Create a pandas DataFrame object with the data and row labels
df4 = pd.DataFrame(data, index = labels)
# Display the DataFrame object in a more readable format using the IPython display module
print("DataFrame: ")
display(HTML(df4.to_html()))

# Replace missing values in the DataFrame with a specified value (4 in this case)
filldata = df4.fillna(4)
# Print and display the DataFrame object with missing values replaced
print("\n---------------------------------------------------------------------------------")
print("Replace missing values in the DataFrame with a specified value (4 in this case):")
display(HTML(filldata.to_html()))

# Calculate the mean of the 'Age' column
meanAge = df4[['Age']].mean()

# Replace missing values in the 'Age' column with the mean value
filldata1 = df4['Age'].fillna(meanAge.iloc[0])

# Replace the original 'Age' column with the new column that has missing values replaced
df4['Age'] = filldata1

# Print the new 'Age' column with missing values replaced
print("\n---------------------------------------------------------------------------------")
print("New 'Age' column with missing values replaced:")
print(filldata1)

```

```python
"""
    Ejemplo de como quitar informaci√≥n repetida 
"""

# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Import the IPython display module to display the DataFrame object in a more readable format
from IPython.core.display import HTML

# Create a pandas DataFrame object with a single column of integers
df8 = pd.DataFrame({'A': [1, 2, 2, 2, 4, 4, 5, 5, 6, 6, 7, 8, 8]})

# Print the DataFrame object and display it in a more readable format using the IPython display module
print("DataFrame: ")
display(HTML(df8.to_html()))

# Select the rows where the value in column 'A' is different from the previous row's value
r = df8.loc[df8['A'].shift() != df8['A']]

# Print a separator and display the selected rows in a more readable format using the IPython display module
print("\n-----------")
print("DataFrame without repeated Data: ")
display(HTML(r.to_html()))
```

```python
# Import the pandas library and give it the alias "pd"
import pandas as pd 

# Import the IPython display module to display the DataFrame object in a more readable format
from IPython.core.display import HTML

# Define a dictionary with data for the DataFrame
data = {
    'Animal': 
        ['Cat', 'Cat', 'Snake', 'Dog', 'Dog', 'Cat', 'Snake', 'Cat', 'Dog', 'Dog'],
    'Age': 
        [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'Visits':
        [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'Priority':
        ['Yes', 'Yes', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', 'No']
}

# Define a list of strings for the row labels
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Create a pandas DataFrame object with the data and row labels
df6 = pd.DataFrame(data, index = labels)

# Write the DataFrame object to a CSV file
df6.to_csv('../Data/csv/AnimalDF.csv')

# Read the CSV file into a new DataFrame object
df_animal = pd.read_csv('../Data/csv/AnimalDF.csv')

# Display the new DataFrame object in a more readable format using the IPython display module
print("CVS file: ")
display(HTML(df_animal.to_html()))

# Write the DataFrame object to an Excel file
df6.to_excel('../Data/excel/AnimalDF.xlsx', sheet_name='Sheet1')

# Read the Excel file into a new DataFrame object
df_animal2 = pd.read_excel('../Data/excel/AnimalDF.xlsx', 'Sheet1', index_col=None, na_values=['NA'])

# Display the new DataFrame object in a more readable format using the IPython display module
print('\n--------------------------------------------')
print("Excel file: ")
display(HTML(df_animal2.to_html()))
```

## **Vizualizations In Pandas Using Matplotlib:**
-------------------------------
-------------------------------

```python
"""
    Ejemplo de Visualizaci√≥n de Pandas Series Usando Matplotlib
"""

# Import the numpy library and give it the alias "np"
import numpy as np
# Import the pandas library and give it the alias "pd"
import pandas as pd
# Import the matplotlib pyplot library and give it the alias "plt"
import matplotlib.pyplot as plt


# Create a pandas Series object with 50 random numbers and a date range index
ts = pd.Series(np.random.randn(50), index=pd.date_range('today', periods=50))

# Calculate the cumulative sum of the Series object
ts = ts.cumsum()  

# Create a new figure and axes object for the plot
fig, ax = plt.subplots()

# Plot the cumulative sum of the Series object
ax.plot(ts)

# Add labels and title

# Set the x-axis label
plt.xlabel('Date')
# Set the y-axis label
plt.ylabel('Cumulative Sum')
# Set the plot title
plt.title('Random Time Series')

# Rotate and format the x-axis date labels for better readability
fig.autofmt_xdate()

# Display the plot
plt.show()
```

```python
"""
    Ejemplo de Visualizaci√≥n de Pandas DataFrame Usando Matplotlib
"""
# Import the numpy library and give it the alias "np"
import numpy as np
# Import the pandas library and give it the alias "pd"
import pandas as pd
# Import the matplotlib pyplot library and give it the alias "plt"
import matplotlib.pyplot as plt

# Generate a DataFrame with random data
df = pd.DataFrame(np.random.randn(50, 4), index=pd.date_range('today', periods=50),
                  columns=['A', 'B', 'X', 'Y'])

df = df.cumsum()

# Create a figure and axes object
fig, ax = plt.subplots()

# Iterate over the columns and plot each one
for column in df.columns:
    ax.plot(df.index, df[column], label=column)

# Add a legend
ax.legend()

# Add labels and title
# Set the x-axis label
plt.xlabel('Date')
# Set the y-axis label
plt.ylabel('Cumulative Sum')
# Set the plot title
plt.title('Random Time DataFrame')

# Rotate date labels on the x-axis
fig.autofmt_xdate()

# Display the plot
plt.show()
```
